```python
#loads data from csv file
def load_data(file_path):
    #reutrns data as a DataFrame
    return pandas.read_csv(file_path)
```

```python
#drop unnessary columns from data for plotting purposes
def data_manipulation(df):
    return df.drop(columns = ['bmi', 'children', 'region', 'sex',], axis = 'columns')
```

```python
#plots relevant data from data set in a scatter graph by age and charges
def plot_charges_by_age_and_smoking(data_frame):    
    # Plotting
    plt.figure(figsize=(10, 6))

    # Plot charges for smokers
    smokers_data = df[df['smoker'] == 'yes']
    plt.scatter(smokers_data['age'], smokers_data['charges'], label='Smokers', alpha=0.7)

    # Plot charges for non-smokers
    non_smokers_data = df[df['smoker'] == 'no']
    plt.scatter(non_smokers_data['age'], non_smokers_data['charges'], label='Non-Smokers', alpha=0.7)

    plt.title('Medical Charges by Age and Smoking Status')
    plt.xlabel('Age')
    plt.ylabel('Medical Charges')
    plt.legend()
    plt.grid(True)
    plt.show()
```
